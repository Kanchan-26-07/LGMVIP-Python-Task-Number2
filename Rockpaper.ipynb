```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to handle user choice
def user_choice(choice):
    choices = ["Rock", "Paper", "Scissors"]
    computer_choice = random.choice(choices)
    result = determine_winner(choice, computer_choice)
    display_result(choice, computer_choice, result)

# Function to determine the winner
def determine_winner(user, computer):
    if user == computer:
        return "It's a tie!"
    elif (user == "Rock" and computer == "Scissors") or \
         (user == "Paper" and computer == "Rock") or \
         (user == "Scissors" and computer == "Paper"):
        return "You win!"
    else:
        return "You lose!"

# Function to display the result
def display_result(user, computer, result):
    clear_output(wait=True)
    display(widgets.HTML(f"<h3>Your choice: {user}</h3>"))
    display(widgets.HTML(f"<h3>Computer's choice: {computer}</h3>"))
    display(widgets.HTML(f"<h2>{result}</h2>"))
    display_buttons()

# Function to display the buttons
def display_buttons():
    btn_rock = widgets.Button(description="Rock", layout=widgets.Layout(width='auto', height='auto'))
    btn_rock.on_click(lambda x: user_choice("Rock"))
    btn_paper = widgets.Button(description="Paper", layout=widgets.Layout(width='auto', height='auto'))
    btn_paper.on_click(lambda x: user_choice("Paper"))
    btn_scissors = widgets.Button(description="Scissors", layout=widgets.Layout(width='auto', height='auto'))
    btn_scissors.on_click(lambda x: user_choice("Scissors"))

    button_box = widgets.HBox([btn_rock, btn_paper, btn_scissors])
    display(button_box)

# Display the buttons initially
display_buttons()
```
